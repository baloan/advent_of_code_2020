```python
import re
RE = "(?P<lo>\d+)-(?P<hi>\d+)\s(?P<c>.):\s(?P<pwd>.*)"
REX = re.compile(RE)

with open("day2_input.txt") as f:
    lines = [s for s in f]
    cnt = 0
    for l in lines:
        mo = REX.match(l)
        lo, hi, c, pwd = mo.groups()
        # count chars in pwd
        n = pwd.count(c)
        if n >= int(lo) and n <= int(hi):
            cnt += 1
        print(mo.groups(), n)
    print(f"{cnt}/{len(lines)}")
```

```python
import re
RE = "(?P<lo>\d+)-(?P<hi>\d+)\s(?P<c>.):\s(?P<pwd>.*)"
REX = re.compile(RE)

with open("day2_input.txt") as f:
    lines = [s for s in f]
    cnt = 0
    for l in lines:
        mo = REX.match(l)
        lo, hi, c, pwd = mo.groups()
        # count chars in pwd
        c0 = pwd[int(lo)-1]
        c1 = pwd[int(hi)-1]
        ok = (c0 == c) != (c1 == c)
        if ok:
            cnt += 1
        print(mo.groupdict(), ok)
    print(f"{cnt}/{len(lines)}")
```
